```python
def calculator_tool(query):
    try:
        return str(eval(query))
    except:
        return "Error in calculation"

def word_count_tool(text):
    return f"Word count: {len(text.split())}"
```

```python
def agent(query):
    # Tool 1: Calculator
    if any(op in query for op in ["+", "-", "*", "/"]):
        return calculator_tool(query)

    # Tool 2: Word Counter
    elif "count words" in query.lower():
        return word_count_tool(query)

    # Simulated LLM responses
    elif "artificial intelligence" in query.lower():
        return "Artificial Intelligence is the simulation of human intelligence in machines that are programmed to think and learn."

    elif "machine learning" in query.lower():
        return "Machine Learning is a subset of AI that enables systems to learn from data."

    else:
        return "This is a general response generated by the intelligent agent."
```

```python
print(agent("25 * 4 + 10"))
print(agent("count words in hello world from AI"))
print(agent("What is artificial intelligence?"))
```

```python
while True:
    q = input("Ask: ")
    if q == "exit":
        break
    print("Agent:", agent(q))
```
